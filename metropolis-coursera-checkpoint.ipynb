# Problem set-up

**Data**: percentage of total change in personnel from last year to this year for 10 companies. 

$y_{i}$: (% of change in company i)

   $y_{i} \ $$\stackrel{i.i.d}\sim \mathcal{N} (\mu , 1)$ 
   
   prior distribution: 
   
   $\mu \sim \ \mathcal{t} (\,0,1,1)$
   

this model is not conjugate. $\longrightarrow$ posterior does not have a known form.

To get posterior sample we are going to set up a markov chain whose stationary distribution is the posterior distribution.



\begin{equation*}
 \mathcal{p} \, ( \mu  | \mathcal{y_{1}} \cdots , \mathcal{y_{n}})  \propto \exp \,[ \frac {n \,(\bar{y}\mu - \mu^2/2) } {1+\mu^2}]
\label{eq:g} \tag{1}
\end{equation*}

\eqref{eq:g} is going to be our $\mathcal{g(\mu)}$

we will use $\log$ of $likelihood$ because it is numrtically more stable. *likelihood* is the product of many potentially small values (hence they are all probabilities) and the computer might treat them az $0$

\begin{equation*}
\log(g(\mu))  \propto (n \,(\bar{y}\mu - \mu^2/2) - \log (1+\mu^2)
\label{eq:lg} \tag{2}
\end{equation*}



## Step 1

write $\eqref{eq:lg}$ in `R`

```r
lg = function(mu, n, ybar){
    
    mu2 = mu^2
    n*(ybar*mu - mu2/2.0) - log(1+mu2)
}
```

now lets write Metropolis-Hastings Sampler with normal proposal distribution

```r
mh = function(n, ybar, n_iter, mu_init, cand_sd){
    #step1
    mu_out = numeric(n_iter)
    accpt = 0 # to keep track of how many candidate we have accepted
    mu_now = mu_init
    lg_now = lg(mu= mu_now, n=n, ybar = ybar)
    
    #step2
    for(i in 1:n_iter){
        mu_cand = rnorm(1,mean=mu_now,sd = cand_sd)
        lg_cand = lg(mu = mu_cand,n=n, ybar = ybar)
        # now we compute acceptance ratio(alpha)
        # Normal distr. is symmetric hence the q() dirmtr. cancells out
        # alpha = g(mu_cand)/g(mu_now)
        lalpha = lg_cand - lg_now
        alpha = exp(lalpha)
        
    #step3
    
        #with the probability alpha we are going to accept the candidate
         u = runif(1)
        if( u< alpha){
            mu_now = mu_cand
            accpt = accpt + 1
            lg_now = lg_cand
        }
        mu_out[i] = mu_now
    }
    list(mu=mu_out, accpt= accpt/n_iter)
}
```

Now, let’s set up the problem.



```r
y = c(1.2, 1.4, -0.5, 0.3, 0.9, 2.3, 1.0, 0.1, 1.3, 1.9)
ybar = mean(y)
```

```r
ybar
```

```r
n = length(y) #10
```

```r
hist(y,freq = FALSE, col='pink',xlim=c(-1,3)) # histogram of the data
curve(dt(x = x,df = 1),lty=2,add = TRUE,col='darkblue') # prior for mu
points(y,rep(0,n),col='green',pch=19) # individual data points
points(ybar,0,pch=15,col='maroon')  # sample mean
```

Finally, we’re ready to run the sampler! Let’s use $ \mathcal{m}=1000 $ iterations and proposal standard deviation (which controls the proposal step size) 3.0, and initial value at the prior median 0

```r
set.seed(43) # set the random seed for reproducibility
post = mh(n=n, ybar=ybar, n_iter=1e3, mu_init=0.0, cand_sd=3.0)
str(post)
```

now let's see what happens if our initial mu is to off!

```r
post = mh(n=n, ybar=ybar, n_iter=1e3, mu_init=30.0, cand_sd=0.9)
post$accpt
```

```r
traceplot(as.mcmc(post$mu),col='red')
```

It took awhile to find the stationary distributio. If we discard the first 150 or so values, it appears like the rest of the samples come from the stationary distribution, our posterior distribution! Let’s plot the posterior density against the prior to see how the data updated our belief about $\mu$

```r
post_keep = post$mu[-c(1,150)]# discard the first 150 samples
```

```r
traceplot(as.mcmc(post$mu_keep),col='violet')
```

```r
plot(density(post$mu_keep, adjust=2.0), main="density estimate of the posterior and prior" , xlim=c(-1.0, 3.0), xlab=expression(mu)) # plot density estimate of the posterior
curve(dt(x=x, df=1), lty=2, add=TRUE,col='darkgreen') # prior for mu
points(ybar,0,pch=15,col='maroon')  # sample mean
curve(0.017*exp(lg(mu=x, n=n, ybar=ybar)), from=-1.0, to=3.0, add=TRUE, col="blue") # approximation to the true posterior in blue
```
